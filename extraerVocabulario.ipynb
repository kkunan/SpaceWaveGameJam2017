```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

import requests
from bs4 import BeautifulSoup


def tomaTxtDeHTML(url):
  r = requests.get(url)
  html = r.content
  #etqHtml = BeautifulSoup(html, "lxml")
  etqHtml = BeautifulSoup(html, features="html.parser")

# Quita estilos y javascript
  for script in etqHtml(["script", "style"]):
    script.extract()
# toma el contenido restante
  textoHTML = etqHtml.get_text()
# crea líneas y quita espacios
  lineas = (linea.strip() for linea in textoHTML.splitlines())
# comprime multilineas
  cachos = (frase.strip() for linea in lineas for frase in linea.split("  "))
# quita líneas en blanco
  texto = '\n'.join(cachos for cachos in cachos if cachos)

  return texto

# encontrar palabras tipo sustantivo, puede probarse con muchas otras cosas, no es perfecta la salida
def dameSustantivos(texto):
  es_sustantivo = lambda pos: pos[:2] == 'NN'
# usar el natural language toolkit
  limpios = nltk.word_tokenize(texto)
  sustantivos = [pal for (pal, pos) in nltk.pos_tag(limpios) if es_sustantivo(pos)] 
  return sustantivos

def imprimePals(pals):
  for p in pals:
    print(p)

# Obtiene una impresión de las palabras de las URL contenidas en el archivo arch
# Falta almacenar estas palabras en otro archivo, ya sea juntas o seaparas por cada archivo
# Definir los archivos a revisar en el archivo noticiasHTML.txt
# Formato: archivo de texto con las ligas de los archivos html uno en cada línea, ej:
#https://www.milenio.com/policia/zapopan-fiscalia-investiga-asesinato-mujer-trans
#https://www.milenio.com/politica/comunidad/mujeres-trans-detenidas-leon-recibieron-golpes-asegura-sspl 
#https://www.milenio.com/politica/comunidad/policias-de-leon-son-investigados-por-agredir-a-mujeres-transexuales
# ...
# Este archivo debe subirse aquí cada que se inicie la sesión
#

def procesaArchURL(arch):
  a = open(arch, 'r')
  for url in a:
    print("Procesando arhivo de la URL: ",url)
    texto = tomaTxtDeHTML(url)
    palabras = nltk.word_tokenize(texto)
    imprimePals(palabras)


#
# EJemplo con una url determinada
#
url = 'https://www.milenio.com/policia/zapopan-fiscalia-investiga-asesinato-mujer-trans'

texto = tomaTxtDeHTML(url)
# Obtener el vocabulario (todas las palabras que aquí ocurren), hay que filtrar muchas palabras "basura"
palabras = nltk.word_tokenize(texto)
imprimePals(palabras)

# Descomentar para:
#   Obtener sustantivos:
# sustantivos = dameSustantivos(texto)
#   Ver los sustantivos hallados:
# imprimePals(sustantivos)

# Ejemplo con todos los archivos en noticiasHTML.txt (debe cargarse en cada sesión)
procesaArchURL("noticiasHTML.txt")
```
